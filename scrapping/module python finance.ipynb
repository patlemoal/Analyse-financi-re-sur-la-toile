```python
import yfinance as yf
```

```python
#définir le ticker
cap = 'CAP.PA'

#récupérer les datas
data = yf.Ticker(cap)

#récupérer les prix (périodicité =1d correspond à un jour, start :date du début, end :date de fin)
datadf = data.history(period='1d', start = '2021-1-1', end = '2021-3-22')

#Afficher
datadf
```

```python
import matplotlib.pyplot as plt
datadf['Close'].plot()
plt.show()
```
